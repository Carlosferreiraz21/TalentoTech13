```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo os horários do dia (0h a 24h)
horas = np.arange(0, 25)

# Criando uma curva fictícia para representar a variação da temperatura ao longo do dia
# A temperatura sobe até o meio-dia e depois começa a cair
temperaturas = np.piecewise(horas, 
                            [horas <= 12, horas > 12], 
                            [lambda x: 15 + (x * 1.25), lambda x: 30 - ((x - 12) * 1.0)])

# Criando o gráfico
plt.figure(figsize=(10, 5))
plt.plot(horas, temperaturas, marker='o', linestyle='-', color='red', label="Temperatura (°C)")

# Adicionando título e rótulos
plt.title("Evolução da Temperatura ao Longo do Dia")
plt.xlabel("Horário (h)")
plt.ylabel("Temperatura (°C)")
plt.xticks(horas)  # Mostra todas as horas no eixo X
plt.grid(True)  # Adiciona grade para facilitar leitura

# Exibir legenda
plt.legend()

# Mostrar o gráfico
plt.show()
```
